```python
import requests
from bs4 import BeautifulSoup
```

```python
# Prompt the user to enter the product to search for
product_name = input("Enter the product to search for: ")

# Format the product name for use in the URL
product_url = product_name.replace(" ", "+")

# Define the URL of the page to scrape
url = f"https://www.flipkart.com/search?q={product_url}"

# Send a request to the page and store the response
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find all the products on the page
products = soup.find_all("div", {"class": "_2kHMtA"})

# Loop through each product and extract its details
for product in products:
    # Get the product title
    title = product.find("div", {"class": "_4rR01T"}).text.strip()

    # Get the product price
    price = product.find("div", {"class": "_30jeq3 _1_WHN1"}).text.strip()

    # Get the product rating
    rating_elem = product.find("div", {"class": "_3LWZlK"})
    rating = rating_elem.text.strip() if rating_elem else "No rating"

    # Print the details for the current product
    print(f"Title: {title}")
    print(f"Price: {price}")
    print(f"Rating: {rating}")
    print()
```
